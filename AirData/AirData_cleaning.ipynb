```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

location = ['Chaozhou', 'Pingdong']
```

```python
file_name = "2017_{}_20180309.xls".format(location[1])
data = pd.read_excel(file_name, sheet_name=None)
data = data['Sheet1']
print(type(data))
data.head()
```

```python
data1 = data.drop(['測站'], axis=1)
data2 = data1.set_index(['日期', '測項'])
data2.head()
```

```python
data_T = data2.transpose()
data_T.head(5)
```

```python
data_T.index.values
```

```python
data_T.columns.values
```

```python
data3 = data_T
# .swaplevel(0, 1, 1)
data3.head()
```

```python
data4 = data3.stack(level=0, dropna=False) # Do not drop missing value
# for hour in data_T.index.values:
#     print(len(data4.loc[[hour]]))
data4.head()
```

```python
print(data4.index.get_level_values(0))
print(type(data4.index.levels[0]))
idx = ([int(t) for t in data4.index.levels[0]])
data4.index = data4.index.set_levels(idx, level=0)
print(data4.index.get_level_values(0))
# data4.columns.values
```

```python
data5 = data4
data5.index.values
data5 = data4.sort_index(level=1, axis=0)
# print(data5.index.values[0][1])
# print(data5.index.values[1][1])
# if (data5.index.values[0][0]) < (data5.index.values[1][0]):
#     print('True')
data5.head()
```

```python
import datetime
# data5.index.get_level_values(1)
# data5.index = data4.index.set_levels(data4.index.get_level_values(1).map(lambda x: pd.to_datetime(x, format='%Y/%m/%d %H:%M')), level=1)
# data5.index.get_level_values(1)
Date = list()
for idx in data5.index.values:
#     print(idx[0], idx[1])
    day = ([int(i) for i in idx[1].split('/')])
#     print(day)
    Date.append(datetime.datetime(day[0], day[1], day[2], idx[0]).strftime('%m/%d/%Y %H:%M'))
print(len(Date))
data5 = data5.reset_index()
data5 = data5.drop(columns=['日期', 'level_0'])
```

```python
data5["Date"] = Date
data5 = data5.set_index("Date")
data5.head()
# data5.index.values
```

```python
data5.to_csv("Pingdong_2017.csv")
```
