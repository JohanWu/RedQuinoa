```python
# -*- encoding: utf8-*-

from datetime import timedelta, date
from bs4 import BeautifulSoup
import requests
import pandas as pd
## $conda install bs4
## $conda install pandas
## $conda install lxml


start_date = date(2017, 1, 1)
end_date = date(2018, 4, 16)

def daterange(start_date, end_date):
#     print((end_date-start_date).days)
    for n in range(int((end_date-start_date).days)):
        yield start_date + timedelta(n)

def scraper(date):
    page = requests.get('http://e-service.cwb.gov.tw/HistoryDataQuery/DayDataController.do?command=viewMain&station=C0R150&stname=%25E4%25B8%2589%25E5%259C%25B0%25E9%2596%2580&datepicker={}'.format(date))
    print("status code:", page.status_code) # page.content
#     print(page.encoding)
    soup = BeautifulSoup(page.content, 'html.parser')
#     with open("%s.html" %date, "w") as f:
#         f.write(str(soup.prettify()))
#         f.close()
    
    ## Each html has two tables. The second one is wanted.
    ## Get the header of the second table.
    second_tr = soup.find(class_='second_tr')
#     print(type(second_tr))
    labels = second_tr.get_text().strip().split('\n')
#     print(labels)
    df = pd.DataFrame(columns=labels)
    ## Get the values of the second table.
    table_body = soup.find_all('table')[1]
    rows = table_body.find_all('tr')
    for i, row in enumerate(rows[2:]):
#         print(row)
        cols = row.find_all('td')
        cols = [x.text.strip() for x in cols]
#         print(cols)
        df.loc[i] = cols
    ## save the DataFrame to csv file
    df.to_csv('./csv/%s.csv' %date)
```

```python
for Date in daterange(start_date, end_date):
    print(Date)
    scraper(Date)
```
