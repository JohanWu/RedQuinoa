```python
# -*- coding: 'utf-8' -*-
import pandas as pd

year = 2017
loc = 0

station = ['Fangliao', 'Chunri']

file_name = '{}_{}_weather.csv'.format(year, station[loc])
```

```python
df = pd.read_csv(file_name, encoding='utf-8')
df.head()
```

```python
list(df)
cols = ['idx', 'Time', 'StnPres', 'SeaPres', 'Temperature', 'Td_dew_point', 'RH', 'WS', 'WD', 'WSGust', 'WDGust', 'Precp', 'PrecpHour', 'Sunshine', 'GloblRad', 'Visb']
df.columns = cols
df.head()
```

```python
df.Time = pd.to_datetime(df.Time)
```

```python
df.set_index(df.Time, inplace=True)
df.head()
```

```python
df = df.apply(pd.to_numeric, errors='coerce')
df.head()
```

```python
df = df.drop(['idx', 'Time'], axis=1)
df.head()
```
