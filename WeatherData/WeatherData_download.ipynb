### Download the Climate Data from CWB
http://e-service.cwb.gov.tw/HistoryDataQuery/index.jsp

```python
# -*- encoding: utf8-*-

from datetime import timedelta, date, datetime
from bs4 import BeautifulSoup
import requests
import pandas as pd
## $conda install bs4
## $conda install pandas
## $conda install lxml

year = 2017
start_date = date(year, 1, 1)
end_date = date(year+1, 1, 1)

loc = 1 # choose the station in which you want to dowload the data
STATION = ['Fangliao', 'Chunri']
URL = [
    'http://e-service.cwb.gov.tw/HistoryDataQuery/DayDataController.do?command=viewMain&station=C0R380&stname=%25E6%259E%258B%25E5%25AF%25AE&datepicker=',
    'http://e-service.cwb.gov.tw/HistoryDataQuery/DayDataController.do?command=viewMain&station=C0R260&stname=%25E6%2598%25A5%25E6%2597%25A5&datepicker='
]
```

```python
def daterange(start_date, end_date):
#     print((end_date-start_date).days)
    for n in range(int((end_date-start_date).days)):
        yield start_date + timedelta(n)
```

```python
def scraper(station, date):
    '''Get data from CWB website.
    
    Parameters
    ===========
    station: URL for the particular station
    date: the specific date
    '''       
#     page = requests.get(station+str(date))
#     print(page.encoding)
    df_list = pd.read_html(station+str(date), encoding='utf-8')
    cols_name = df_list[1].iloc[1]
#     print(cols_name)
    df = df_list[1][2:] # only take the value components in the dataframe
#     df.columns = cols_name
    
    dt = [int(i) for i in str(date).split('-')]
    # print(dt)
    dt = datetime(dt[0], dt[1], dt[2]) # let datetime object have hours and mins as well
    Date = list()
    for h in df[0]:
        Date.append((dt+timedelta(hours=int(h)-1)).strftime('%Y-%m-%d %H:%M'))
    
    df.iloc[:, 0] = Date

    return df, cols_name
```

```python
DF = pd.DataFrame()

for Date in daterange(start_date, end_date):
    print(Date)
    df, cols_name = scraper(URL[loc], Date)
    DF = DF.append(df, ignore_index=True)
DF.columns = cols_name
DF.tail()
```

```python
DF.to_csv("{}_{}_weather.csv".format(year, STATION[loc]), encoding='utf-8')
```
